```python
import re
```

```python
# Читаем текст из нашей книжки (использую книгу по шахматам)
text = ""
with open('drive/MyDrive/LLM_RAG/text.txt', 'r') as f:
  lines = [line.strip() for line in f.readlines()]
  text = ' '.join(lines)
preprocessed_text = re.sub(r'[.?!]+', '.', text)
len(preprocessed_text)
```

```python
# Разобьем на сегменты (в сегменте по 3 предложению)
sentences = preprocessed_text.split('.')
segments = []
i = 0
while i < len(sentences):
  segments.append(' '.join(sentences[i:i+3]))
  i += 3
len(segments), len(sentences), max([len(segment) for segment in segments])
```

```python
# Применим модель к сегментам
input_texts = ['passage: ' + segment for segment in segments]
embeddings = model.encode(input_texts, normalize_embeddings=True)
```

```python
import numpy as np
with open('drive/MyDrive/LLM_RAG/embeddings.npy', 'wb') as f:
  np.save(f, embeddings)
with open('drive/MyDrive/LLM_RAG/segments.txt', 'w') as f:
  for segment in segments:
    f.write(segment + '[/SEP\]')
```

```python
import numpy as np
from scipy.spatial.distance import cosine, euclidean, correlation
from scipy.spatial.distance import cdist
class MyKNeighbors:
    def __init__(self, n_neighbors):
        self.n_neighbors = n_neighbors
        with open('drive/MyDrive/LLM_RAG/embeddings.npy', 'rb') as f:
            self.X = np.load(f)
        with open('drive/MyDrive/LLM_RAG/segments.txt', 'r') as f:
            self.segments = f.read().strip().split('[/SEP\]')[:-1]

    def predict(self, queries):
        queries = ['query: ' + query for query in queries]
        embeddings = model.encode(queries, normalize_embeddings=True)
        ind = np.argpartition(cdist(embeddings, self.X, metric='cosine'), self.n_neighbors, axis=1)[:,:self.n_neighbors]
        most_appropriate_segments = []
        for indices in ind:
          segs = [self.segments[i] for i in indices]
          most_appropriate_segments.append(segs)
        return most_appropriate_segments
```

```python
kneigh = MyKNeighbors(10)
```

```python
predicted_segments = kneigh.predict(
    ['какие дома существуют в столице',
     'что местные думают о приезжих']
)
```

```python
len(predicted_segments), len(predicted_segments[0])
```

```python
# Коррелирующие сегменты для первого предложения
# какие дома существуют в столице
for i, segment in enumerate(predicted_segments[0]):
  print(f'{i+1})', segment, end='\n\n')
```

```python
# Коррелирующие сегменты для второго предложения
# что местные думают о приезжих
for i, segment in enumerate(predicted_segments[1]):
  print(f'{i+1})', segment, end='\n\n')
```
