```python
from transformers import AutoTokenizer, AutoModelForCausalLM, AutoConfig

model_name_or_path = 'openchat/openchat-3.5-1210'
tokenizer = AutoTokenizer.from_pretrained(model_name_or_path)
base_model_config = AutoConfig.from_pretrained(model_name_or_path)

model = AutoModelForCausalLM.from_pretrained(
    model_name_or_path,
    torch_dtype=base_model_config.torch_dtype,
    load_in_8bit=True,
    device_map="auto",
)
model.eval()
```

```python
def generate(prompt, generation_config):
    data = tokenizer(
        prompt,
        return_tensors="pt",
    )
    data = {k: v.to(model.device) for k, v in data.items()}
    output_ids = model.generate(
        **data,
        generation_config=generation_config
    )
    outputs = []
    for sample_output_ids, sample_input_ids in zip(output_ids, data["input_ids"]):
        sample_output_ids = sample_output_ids[len(sample_input_ids):]
        sample_output = tokenizer.decode(sample_output_ids, skip_special_tokens=True)
        sample_output = sample_output.replace("</s>", "").strip()
        outputs.append(sample_output)
    return outputs[0]
```

## Zero-shot prediction

```python
from transformers import GenerationConfig
generation_config = GenerationConfig.from_pretrained(model_name_or_path)
generation_config
```

```python
generation_config.do_sample = True
generation_config.max_new_tokens = 128
generation_config.temperature = 0.05
generation_config.top_k = 50
generation_config.top_p = 0.8
generation_config.num_beams = 1
generation_config
```

```python
from tqdm import tqdm
class ZeroShotPredictor:
  question: str

  def __init__(self, question: str, question_op_word: str, conf):
    self.question = question
    self.qopw = question_op_word
    self.conf = conf

  def predict(self, samples: list[tuple[str, str]]):
    messages = []
    messages2 = []
    for (sentence, entity) in samples:
      messages.append({'role': 'user', 'content': self.question.format(entity, sentence)})
      messages2.append({'role': 'user', 'content': self.qopw.format(entity, sentence)})
    prompts = tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
    prompts2 = tokenizer.apply_chat_template(messages2, tokenize=False, add_generation_prompt=True)
    results = generate(prompts, self.conf)
    results2 = generate(prompts2, self.conf)
    return [results, results2]
```

```python
template = 'Необходимо определить тональность в предложении по отношению к сущности {}. Тональность может быть только позитивной или негативной. Напиши тональность в предложении одним словом. Больше ничего писать не нужно. Предложение: {}'
template2 = 'Необходимо определить слово, выражающее в предложении по отношению к сущности {}. Процитируй слово или словосочетание, выражающее тональность к сущности в предложении. Больше ничего писать не нужно. Предложение: {}'
clf0 = ZeroShotPredictor(template, template2, generation_config)
```

```python
inp = list(zip(val['sentence'].to_list(), val['entity'].to_list()))
```

```python
preds = []
for elem in tqdm(inp):
  preds.append(clf0.predict([elem]))
preds
```

```python
import numpy as np
labels = []
for ans in preds:
  if ans[0].startswith("негатив"):
    labels.append(-1)
  else:
    labels.append(1)
labels = np.array(labels)
(val['label'].to_numpy() == labels).sum() / labels.shape  # 0.88
```

```python
(val['label'].to_numpy() == labels)
```

```python
from tqdm import tqdm
class FewShotPredictor:
  question: str

  def __init__(self, question: str, question_op_word: str, conf, example_sentence, example, example_op_word):
    self.exs = example_sentence
    self.question = question
    self.ex = example
    self.qopw = question_op_word
    self.exopw = example_op_word
    self.conf = conf

  def predict(self, samples: list[tuple[str, str]]):
    messages = []
    messages2 = []
    for (sentence, entity) in samples:
      messages.append({'role': 'user', 'content': self.question.format(entity, sentence, self.exs, self.ex)})
      messages2.append({'role': 'user', 'content': self.qopw.format(entity, sentence, self.exs, self.exopw)})
    prompts = tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
    prompts2 = tokenizer.apply_chat_template(messages2, tokenize=False, add_generation_prompt=True)
    results = generate(prompts, self.conf)
    results2 = generate(prompts2, self.conf)
    return [results, results2]
```

```python
template = 'Необходимо определить тональность в предложении по отношению к сущности {}. Тональность может быть только позитивной или негативной. Напиши тональность в предложении одним словом. Больше ничего писать не нужно. Предложение: {}. Пример: для предложения {} ответ {}.'
template2 = 'Необходимо определить слово, выражающее в предложении по отношению к сущности {}. Процитируй слово или словосочетание, выражающее тональность к сущности в предложении. Больше ничего писать не нужно. Предложение: {}. Пример: для предложения {} ответ {}.'
clf1 = FewShotPredictor(template, template2, generation_config, 'Речь идет о скандальном деле Mediaset, в рамках которого Берлускони обвинялся в масштабных финансовых махинациях.', 'негативная', "скандальном")
```

```python
inp = list(zip(val['sentence'].to_list(), val['entity'].to_list()))
```

```python
preds = []
for elem in tqdm(inp):
  preds.append(clf1.predict([elem]))
preds
```
